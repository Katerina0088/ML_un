3\. Дана функция $f(x)$. Найдите (численно) производную данной функции $f'(x)$ на полуинтервале (0, 10]. Постройте в одной системе координат график $f(x)$, $f'(x)$ и $g(x) = 0$. Изобразите графики различными цветами и включите сетку. 

$$f(x) = \frac{sin(x)}{\ln(x) + 1}$$

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return torch.sin(x) / (torch.log(x) + 1)


def numerical_derivative(f, x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)


def torch_derivative(f, x):
    x_tensor = torch.tensor(x, requires_grad=True)
    y = f(x_tensor)
    y.backward() 
    return x_tensor.grad.item()


x_values = np.linspace(0.01, 100, 100)  


f_values = [f(torch.tensor(x)).item() for x in x_values]


f_prime_values = [numerical_derivative(f, torch.tensor(x)).item() for x in x_values]


f_prime_torch_values = [torch_derivative(f, x) for x in x_values]


plt.figure(figsize=(10, 6))
plt.plot(x_values, f_values, label='$f(x)$', color='blue')
plt.plot(x_values, f_prime_values, label='$f\'(x)$ (численно)', color='green')
plt.plot(x_values, f_prime_torch_values, label='$f\'(x)$ (torch)', color='red', linestyle='--')
plt.axhline(0, color='black', linestyle='--', label='$g(x) = 0$')
plt.xlabel('x')
plt.ylabel('Значения')
plt.title('Графики $f(x)$, $f\'(x)$ и $g(x) = 0$')
plt.grid(True)
plt.legend()
plt.show()
```

4\. Найдите локальный минимум функции $f(x)$ при помощи метода градиетного спуска. В качестве начальной точки используйте $x_0 = 4$. Найдите локальный максимум этой же функции, используя в качестве начальной точки $x_0'=9$.

$$f(x) = \frac{sin(x)}{\ln(x) + 1}$$

```python
import torch


def f(x):
    return torch.sin(x) / (torch.log(x) + 1)


def gradient_descent(start_x, learning_rate=0.01, max_iters=1000, tol=1e-6):
    x = torch.tensor(start_x, dtype=torch.float32, requires_grad=True)  
    for i in range(max_iters):
        y = f(x) 

        y.backward()  
        grad = x.grad.item()
        
        
        
        x_new = x - learning_rate * grad
        
       
        if torch.abs(x_new - x) < tol:
            break
        
       
        x.data = x_new
        x.grad.zero_()  
    
    return x.item(), y.item()


def gradient_ascent(start_x, learning_rate=0.01, max_iters=1000, tol=1e-6):
    x = torch.tensor(start_x, dtype=torch.float32, requires_grad=True)  
    for i in range(max_iters):
        y = f(x)
        y.backward()  
        grad = x.grad.item()
        
     
        x_new = x + learning_rate * grad
        
     
        if torch.abs(x_new - x) < tol:
            break
        
       
        x.data = x_new
        x.grad.zero_() 
    
    return x.item(), y.item()


x_min, f_min = gradient_descent(start_x=4)
print(f"Локальный минимум: x = {x_min}, f(x) = {f_min}")


x_max, f_max = gradient_ascent(start_x=9)
print(f"Локальный максимум: x = {x_max}, f(x) = {f_max}")
```
